## Users and Tweets

```python
import os
import pandas as pd

def get_number_of_rows(file):
    return sum(1 for line in open(file))

genuine_accounts_tweets = get_number_of_rows('genuine_accounts.csv/tweets.csv')
genuine_accounts_users = get_number_of_rows('genuine_accounts.csv/users.csv')
fake_followers_tweets = get_number_of_rows('fake_followers.csv/tweets.csv')
fake_followers_users = get_number_of_rows('fake_followers.csv/users.csv')
social_spambots_1_tweets = get_number_of_rows('social_spambots_1.csv/tweets.csv')
social_spambots_1_users = get_number_of_rows('social_spambots_1.csv/users.csv')
social_spambots_2_tweets = get_number_of_rows('social_spambots_2.csv/tweets.csv')
social_spambots_2_users = get_number_of_rows('social_spambots_2.csv/users.csv')
social_spambots_3_tweets = get_number_of_rows('social_spambots_3.csv/tweets.csv')
social_spambots_3_users = get_number_of_rows('social_spambots_3.csv/users.csv')
traditional_spambots_1_tweets = get_number_of_rows('traditional_spambots_1.csv/tweets.csv')
traditional_spambots_1_users = get_number_of_rows('traditional_spambots_1.csv/users.csv')
traditional_spambots_2_users = get_number_of_rows('traditional_spambots_2.csv/users.csv')
traditional_spambots_3_users = get_number_of_rows('traditional_spambots_3.csv/users.csv')
traditional_spambots_4_users = get_number_of_rows('traditional_spambots_4.csv/users.csv')

print(f'genuine_accounts_tweets: {genuine_accounts_tweets}')
print(f'genuine_accounts_users: {genuine_accounts_users}')
print(f'fake_followers_tweets: {fake_followers_tweets}')
print(f'fake_followers_users: {fake_followers_users}')
print(f'social_spambots_1_tweets: {social_spambots_1_tweets}')
print(f'social_spambots_1_users: {social_spambots_1_users}')
print(f'social_spambots_2_tweets: {social_spambots_2_tweets}')
print(f'social_spambots_2_users: {social_spambots_2_users}')
print(f'social_spambots_3_tweets: {social_spambots_3_tweets}')
print(f'social_spambots_3_users: {social_spambots_3_users}')
print(f'traditional_spambots_1_tweets: {traditional_spambots_1_tweets}')
print(f'traditional_spambots_1_users: {traditional_spambots_1_users}')
print(f'traditional_spambots_2_users: {traditional_spambots_2_users}')
print(f'traditional_spambots_3_users: {traditional_spambots_3_users}')
print(f'traditional_spambots_4_users: {traditional_spambots_4_users}')
```
